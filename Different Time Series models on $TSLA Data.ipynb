## Simple Moving Average

```python
df_tesla['Open'].plot(figsize=(20,9))
```

```python
df_tesla['Open: 10 days rolling'] = df_tesla['Open'].rolling(window=10, min_periods=1).mean()
```

```python
df_tesla[['Open','Open: 10 days rolling']].plot(xlim=['2020-01-01','2024-02-29'],figsize=(25,11))
```

```python
df_tesla['Open: 30 days rolling'] = df_tesla['Open'].rolling(window=30, min_periods=1).mean()
```

```python
df_tesla['Open: 50 days rolling'] = df_tesla['Open'].rolling(window=50, min_periods=1).mean()
```

```python
df_tesla[['Open','Open: 10 days rolling','Open: 30 days rolling','Open: 50 days rolling']].plot(xlim=['2021-01-01','2024-02-29'],figsize=(25,11))
```

__Disadvantages of Simple Moving Average:__
* Similar importance is given to all the data. for any window the equal weights are applied and the averages are computed making it inaccurate. (x1+x2+x3+---------+xn)/n
* So that is the reason we use **"Exponential Weighted Moving Average".**

Before learning about EWMA we will have to learn about __Cummulative Moving Average__.


### Cummulative Moving Average

```python
df_tesla['Open'].expanding().mean().plot(figsize=(20,15))
```

### Exponential Moving Average

*EMA(t) = a * (x(t)) + (1-a) * EMA(t-1)*

where a is the weight

This is used to prevent any lags.


```python
#EMA
df_tesla['EMA_0.1'] = df_tesla['Open'].ewm(alpha=0.1, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.1']].plot(xlim=['2020-01-01','2024-02-29'],figsize=(25,11))
```

```python
df_tesla['EMA_0.3'] = df_tesla['Open'].ewm(alpha=0.3, adjust=False).mean()
```

```python
df_tesla[['Open','EMA_0.3']].plot(xlim=['2023-01-01','2024-02-29'],figsize=(25,11))
```

### Exponential Weighted Moving Average

```python
df_tesla['EWMA_0.1'] = df_tesla['Open'].ewm(span=5, adjust=False).mean()
```
